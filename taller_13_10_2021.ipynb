```python
# uso de los tabuladores
# if con else

# asignación de un valor a x
x = 20.01

if x == 20:
    print("El valor de x es 20 ")
    else:
        print("El valor de x NO es igual a 20 ")
```

```python
# uso de los tabuladores
# if con else

# asignación de un valor a x
x = 20.01

if x == 20:
    print("El valor de x es 20 ")
else:
    print("El valor de x NO es igual a 20 ")
```

```python
# if con else

# asignación de un valor a x
x = 20
# ==, !=, <=, >=, and, or

# uso del and y <=, >0 
if x >= 0 and x <= 20 :
    print("El valor de x es 20 ")
else:
    print("El valor de x NO es igual a 20 ")
```

```python
# if con else

# asignación de un valor a x
x = 21
# ==, !=, <=, >=, and, or

print("El valor de x es:", x)
# uso del and y <=, >0 
if ((x >= 0) and (x <= 20)) :
    print("El valor de x esta entre 0 y 20 ")
else:
    print("El valor de x es mayor a 20 ")
```
